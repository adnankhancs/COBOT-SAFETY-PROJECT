```python
# Input and output file paths

def avi_to_mp4(input_file,output_file):
# Open the video file
    cap = cv2.VideoCapture(input_file)

# Get video properties
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create a VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4
    out = cv2.VideoWriter(output_file, fourcc, fps, (width, height))

# Check if CUDA is available
    if cv2.cuda.getCudaEnabledDeviceCount() > 0:
        print("CUDA is available. Using GPU acceleration.")
    # Create a CUDA stream
        stream = cv2.cuda_Stream()
    else:
        print("CUDA is not available. Using CPU.")

# Read and process each frame
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

    # If CUDA is available, upload the frame to GPU
        if cv2.cuda.getCudaEnabledDeviceCount() > 0:
            gpu_frame = cv2.cuda_GpuMat()
            gpu_frame.upload(frame, stream=stream)
        # Perform any GPU-accelerated processing here if needed
            frame = gpu_frame.download(stream=stream)

    # Write the frame to the output file
        out.write(frame)

# Release resources
    cap.release()
    out.release()
    print(f"Video conversion complete. Output saved to {output_file}")
```
